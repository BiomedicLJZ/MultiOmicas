```python
from Bio import SeqIO as SIO
from Bio.Seq import Seq as S
from Bio.SeqRecord import SeqRecord as SR
from tkinter import Tk as TK, filedialog as FD
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
root=TK()
root.withdraw()
root.update()
filename=FD.askopenfilename(title='Select a file',filetypes=[('FASTA files','*.fasta'),('FASTA files','*.fa'),('FASTA files','*.fna'),("All files","*.*")])
```

```python
sequences=[]
for record in SIO.parse(filename,'fasta'):
    sequences.append(record)
```

```python
for i in range(len(sequences)):
    sequences[i].seq=sequences[i].seq.upper()
```
